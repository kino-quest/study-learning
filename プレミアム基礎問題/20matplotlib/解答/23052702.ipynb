```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
%matplotlib inline
```

```python
df = pd.read_excel('SSDSE-E-2023.xlsx', skiprows=2, index_col = '都道府県')
df = df.drop('全国')
df.head()
```

```python
plt.figure(figsize = (10,5))                    #2-4
plt.bar(df.index, df['総人口'])                   #2-1
plt.ylabel('総人口', fontsize = 12)              #2-3   
plt.xticks(rotation = 90)                       #2-5
plt.ticklabel_format(style='plain',axis='y')    #2-2
plt.title('都道府県別総人口', fontsize= 18)        #2-6
plt.grid()                                      #2-7

plt.show()
```

```python
x = df['消費支出（二人以上の世帯）']
y = df['食料費（二人以上の世帯）']
osaka_x = x.loc['大阪府']
osaka_y = y.loc['大阪府']
kanto_list = ['茨城県', '栃木県', '群馬県','埼玉県', '千葉県', '東京都', '神奈川県']
kanto_x = x.loc[kanto_list]
kanto_y = y.loc[kanto_list]

plt.scatter(x, y, color = 'b')                      #3-1
plt.scatter(osaka_x, osaka_y, color = 'g')          #3-3
plt.scatter(kanto_x,kanto_y, color = 'r')           #3-5

plt.annotate('大阪府', xy = (osaka_x, osaka_y))      #3-4

for i in kanto_list:                                  #3-6
    plt.annotate(i, xy = (x.loc[i], y.loc[i]))       
    
plt.xlabel('消費支出（二人以上の世帯）', fontsize = 12)    #3-2
plt.ylabel('食料費（二人以上の世帯）', fontsize = 12)      #3-2
plt.show()
```

```python
fig, axes = plt.subplots(3,1)
axes[0].plot(df.index, df['図書館数'], marker = 'o', label = '図書館数')
axes[1].plot(df.index, df['博物館数'], marker = 'o', label = '博物館数')
axes[2].plot(df.index, df['映画館数'], marker = 'o', label = '映画館数')
axes[0].legend()
axes[1].legend()
axes[2].legend()
plt.show()
```
