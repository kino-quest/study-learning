```python
#1
import pandas as pd

import matplotlib.pyplot as plt

import japanize_matplotlib

%matplotlib inline

df=pd.read_excel('SSDSE-E-2023.xlsx',header=2)
```

```python
#2
df_1=df.loc[8:14,'都道府県':'外国人人口',]
df_1=df_1.sort_values('総人口',ascending=False)
```

```python
#3
X=df_1['都道府県']
Y=df_1['日本人人口']
plt.bar(X,Y,width=0.8)
plt.show()
```

```python
#4
X=df_1['都道府県']
Y=df_1['日本人人口']
plt.barh(X,Y,height=0.5)
plt.show()
```

```python
#5
X=df_1['都道府県']
Y=df_1['日本人人口']
plt.figure(figsize=(10,6))
plt.bar(X,Y,width=0.8)
plt.xticks(rotation=45)
plt.xlabel('県名')
plt.ylabel('人口',rotation=0,fontsize=18)
plt.title('関東の人口')
plt.savefig('bar.png')
plt.show()

```

```python
#7
X=df_1['都道府県']
Y1=df_1['日本人人口']
Y2=df_1['外国人人口']

plt.bar(X,Y1,align='edge',width=0.5,label='日本人')
plt.bar(X,Y2,align='center',width=0.5,label='外国人')

plt.legend()
```

```python
#8-1
X=df_1['都道府県']
Y1=df_1['日本人人口']
Y2=df_1['外国人人口']

plt.bar(X,Y1,label='日本人',color='blue')
plt.bar(X,Y2,label='外国人',color='red',bottom=Y1)

plt.legend()
```

```python
#8-2
X=df_1['都道府県']
Y1=df_1['日本人人口']
Y2=df_1['外国人人口']

plt.barh(X,Y1,label='日本人',color='blue')
plt.barh(X,Y2,label='外国人',color='red',left=Y1)

plt.legend()
```

```python
#9
#問題９　以下の2個のグラフを横に並べて表示してください。
#・各都道府県の '15歳未満人口', '15～64歳人口', '65歳以上人口' の折れ線グラフを１つのグラフにしたもの
#・各都道府県の '15歳未満人口', '15～64歳人口', '65歳以上人口' の積み上げ棒グラフを１つのグラフに表したもの

df_3=df[['都道府県','15歳未満人口', '15～64歳人口', '65歳以上人口']][1:]

x=df_3['都道府県']
y1=df_3['15歳未満人口']
y2=df_3['15～64歳人口']
y3=df_3['65歳以上人口']
```

```python
plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.xticks(rotation=45,fontsize=5)

plt.subplot(1,2,2)
plt.bar(x,y1,label='-15')
plt.bar(x,y2,label='16-64',bottom=y1)
plt.bar(x,y3,label='65-',bottom=y2+y1)
plt.xticks(rotation=45,fontsize=5)
plt.legend()
plt.show()
```
