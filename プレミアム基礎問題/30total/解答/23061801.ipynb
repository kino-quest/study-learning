```python
#1
import math

n=1

while True:
    num_1=int(input("最大公約数が8になるような整数の組み合わせを入力してください"))
    num_2=int(input("最大公約数が8になるような整数の組み合わせを入力してください"))

    if math.gcd(num_1,num_2)==8:
        print('正解です')
        break
    elif n>=5:
        print('失格です')
        break
    else:
        print('やり直し')
        n=n+1
```

```python
#11
df_o=df_m[['都道府県','小学校児童数' ,'中学校生徒数','高等学校生徒数']][1:]
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib
%matplotlib inline
```

```python
#12
x=df_o['都道府県']
y_1=df_o['小学校児童数']
y_2=df_o['中学校生徒数']
y_3=df_o['高等学校生徒数']

plt.figure(figsize=(10,8))

plt.xticks(rotation=90)

plt.bar(x,y_1,label='小学校児童数')
plt.bar(x,y_2,bottom=y_1,label='中学校生徒数')
plt.bar(x,y_3,bottom=y_1+y_2,label='高等学校生徒数')
plt.legend()
plt.show()
```
