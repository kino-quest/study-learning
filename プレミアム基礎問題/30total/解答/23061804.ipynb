```python
#1
import pandas as pd
df=pd.read_excel('SSDSE-E-2023.xlsx',skiprows=2, index_col='地域コード')
df.head()
```

```python
#2
df = df.drop('R00000')
df.head(3)
```

```python
#3
df_a=df.loc[:,'都道府県':'死亡数']
df_a.head()
```

```python
#4
df_a.loc['R01000':'R07000','地域']='北海道東北'
df_a.loc['R08000':'R14000','地域']='関東'
df_a.loc['R15000':'R18000','地域']='北陸'
df_a.loc['R19000':'R23000','地域']='中部'
df_a.loc['R24000':'R30000','地域']='近畿'
df_a.loc['R31000':'R39000','地域']='中国四国'
df_a.loc['R40000':,'地域']='九州沖縄'
df_a.head(10)
```

```python
#5
df_a_tmp1 = df_a[['地域']]
df_a_tmp2 = df_a.loc[:,'都道府県':'死亡数']
df_b = pd.concat([df_a_tmp1,df_a_tmp2], axis=1)
df_b.head(10)
```

```python
#6
df_b=df_b.drop('合計特殊出生率',axis=1)
df_b.head()
```

```python
#7
import matplotlib.pyplot as plt
import japanize_matplotlib
plt.bar(df_b['地域'],df_b['総人口'])
plt.show()
```

```python
#8
df_c=df_b.groupby('地域').sum('総人口').sort_values('総人口',ascending=False)
df_c
```

```python
#9
plt.bar(df_c.index,df_c['総人口'])
plt.title('地域別総人口',fontsize=14)
plt.show()
```

```python
#10
y1=df_c['15歳未満人口']
y2=df_c['15～64歳人口']
y3=df_c['65歳以上人口']

plt.bar(df_c.index, y1, label='15歳未満人口')
plt.bar(df_c.index, y2, label='15～64歳人口',bottom=y1)
plt.bar(df_c.index, y2, label='15～64歳人口',bottom=y1+y2)
plt.title('地域世代別総人口',fontsize=14)
plt.legend()
plt.savefig('地域世代別総人口.png', format='png')
plt.show()
```

```python
#11
z1=df_c['出生数']
z2=df_c['死亡数']

plt.plot(df_c.index, z1, label='出生数')
plt.plot(df_c.index, z2, label='死亡数')
plt.title('地域別_出生数x死亡数',fontsize=14)
plt.legend()
plt.show()
```
