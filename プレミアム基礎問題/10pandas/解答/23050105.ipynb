```python
#問題文に含めない
import pandas as pd
df = pd.read_excel("SSDSE-E-2023.xlsx")
df.head()
```

```python
#1
col_21=df.iloc[0].isin(['年度', 2021])
col_21
```

```python
#2
df_21 = df.loc[:, col_21]
df_21.head()
```

```python
#3
new_header = df_21.iloc[1]
df_21 = df_21[2:]
df_21.columns = new_header
df_21.head()
```

```python
#4
df_21=df_21[df_21['都道府県'] !='全国']
df_21.head()
```

```python
#5
df_21[['都道府県','総人口']].sort_values(by='総人口',ascending=False).head(5)
```

```python
#6
df_21[['都道府県','総人口']].sort_values(by='総人口',ascending=True).head(5)
```

```python
#7
df_21['高齢化率']=df_21['65歳以上人口']/df_21['総人口']*100
```

```python
#8
df_21['高齢化率']=df_21['高齢化率'].apply('{:.1f}'.format)
df_21['高齢化率'].head(5)
```

```python
#9
df_21[['都道府県','総人口','高齢化率']].sort_values(by='高齢化率',ascending=False).head(5)
```

```python
#10
df_21['外国人率']=((df_21['総人口']-df_21['日本人人口'])/df_21['総人口'])*100
```

```python
#11
df_21['外国人率']=df_21['外国人率'].apply('{:.2f}'.format)
```

```python
#12
df_21[['都道府県','総人口','外国人率']].sort_values(by='外国人率',ascending=False).head(5)
```

```python
#13
df_21['エリア'] = df_21['都道府県'].str[-1]
```

```python
#14
df_21[['総人口','エリア']].groupby('エリア').sum()
```

```python
#15
df_21.pivot_table(index=['エリア'],values=['総人口'],aggfunc='sum')
```
