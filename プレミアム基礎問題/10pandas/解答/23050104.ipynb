```python
# 1
df = pd.read_excel('SSDSE-E-2023.xlsx', skiprows = 2, index_col=0)
df = df.drop('R00000')
df.head()
```

```python
# 2
list(df)
# type(list(df))
df.columns
# type(df.columns)
df.columns.get_loc('図書館数')
```

```python
# 3
c01 = df.columns.get_loc('都道府県')
c02 = df.columns.get_loc('小学校児童数')
c03 = df.columns.get_loc('図書館数')
df01 = df.iloc[:, [c01,c02,c03]]
# df01.set_index(df01.columns[0] ,inplace = True)
df01.head()
```

```python
# 4
sg = df01['小学校児童数']
lib = df01['図書館数']
df01['図書館１つ当たりの児童数'] = round(sg/lib).astype(int)
df01 = df01.sort_values('図書館１つ当たりの児童数', ascending = False)
df01.head()
```

```python
# 5
df02_0 = df.copy()
df02_1 = df02_0['都道府県']

# 「従業者数（民営）」を含むカラム名で抽出
columns_working = [column for column in df.columns if '従業者数（民営）' in column]
df02_2 = df.loc[:,columns_working]

# 結合
df02 = pd.merge(df02_1, df02_2, on='地域コード')
df02.head()
```

```python
# 6
for i in columns_working[1:]:
    df02[i] = df02[i] / df02_2[columns_working[0]]
    
df02 = df02.sort_values('従業者数（民営）（宿泊業，飲食サービス業）', ascending=False)
df02[columns_working[1:]] = df02[columns_working[1:]].applymap(lambda x : '{:.2%}'.format(x))
df02.head(10)
```

```python
# 7
c01 = df.columns.get_loc('都道府県')
c04 = df.columns.get_loc('総人口')
c05 = df.columns.get_loc('出生数')
c06 = df.columns.get_loc('死亡数')
c07 = df.columns.get_loc('転入者数（日本人移動者）')
c08 = df.columns.get_loc('転出者数（日本人移動者）')
df03 = df.iloc[:, [c01,c04,c05,c06,c07,c08]]
df03.head()
```

```python
# 8
df03_1 = df03[df03['転入者数（日本人移動者）'] > df03['転出者数（日本人移動者）']]
list(df03_1['都道府県'])
```

```python
# 9
df03['増減数'] = df03['出生数'] + df03['転入者数（日本人移動者）'] - df03['死亡数'] - df03['転出者数（日本人移動者）']
df03.head()
```

```python
# 10
df03_2 = df03.copy()

df03_2['ランク分け'] = ''
for i, row in enumerate(df03_2.itertuples()):
    if row[7]/row[2] > 0:
        df03_2.iloc[i,7] = 'A'
    elif row[7]/row[2] > -0.005:
        df03_2.iloc[i,7] = 'B'
    elif row[7]/row[2] > -0.01:
        df03_2.iloc[i,7] = 'C'
    else:
        df03_2.iloc[i,7] = 'D'
df03_2.sort_values('ランク分け')
df03_2.head()
```

```python
# 11
df03_3 = df03.copy()
df03_3.insert(0,'地方','')
area = ['北海道東北','関東','北陸','中部','近畿','中郷四国','九州沖縄']
df03_3.iloc[0:7, 0] = area[0]
df03_3.iloc[7:14, 0] = area[1]
df03_3.iloc[14:18, 0] = area[2]
df03_3.iloc[18:24, 0] = area[3]
df03_3.iloc[24:30, 0] = area[4]
df03_3.iloc[30:39, 0] = area[5]
df03_3.iloc[39:, 0] = area[6]
df03_3[['地方','増減数']].groupby('地方').mean().round(0).astype(int).reindex(index=area)
```
