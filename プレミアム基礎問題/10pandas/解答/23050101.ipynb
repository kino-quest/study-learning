```python
#問題４
#振りなおした番号を利用して、全国、北海道、東北、関東、中部、関西、四国、中国、九州に地方カテゴリ分けした列を作りましょう。
# 地方カテゴリは地域コードと都道府県の列の間に入れましょう。
df.insert(1,'地方カテゴリ',0) #第一引数：挿入位置　第二引数：列名　第三引数：値
df['地方カテゴリ']=df['地域コード']
df.head(5)
```

```python
#問題５
#isnullを使って欠損値の有無を確認しましょう。
#もし、欠損値があれば平均値を入れましょう。

df.info()

#⇒⇒欠損値なし
```

```python
#問題６　
 #dfに新しい列　"転入超過"を追加し、転入者数が転出者数よりも上回っている都道府県は１、下回っている県には０を入力しましょう。
# ※転入者数（日本人移動者）/ 転出者数（日本人移動者）から算出する

df.insert(7,'転入超過',0) #第一引数：挿入位置　第二引数：列名　第三引数：値

#まずは転入者数ー転出者数の計算結果を転入超過の列に格納する
df['転入超過']=df["転入者数（日本人移動者）"]-df["転出者数（日本人移動者）"]

df.head(5)
```

```python
df['転入超過']=df['転入超過'].map(lambda x:1 if x>=0 else 0)
df.head(5)
```

```python
#問題７
#dfに新しい列を追加して、全面積あたりの耕地面積率を算出したものを入力しましょう。

df['耕地面積率 /%']=round(df['耕地面積']/df['総面積（北方地域及び竹島を除く）']*100).astype(int)
df.head(5)
```

```python
#問題８
#dfのデータの概要をdescribe で把握しましょう。 ※問題９の前にチェックしておく
df.describe(include="all")
```

```python
#問題9
 #dfの幼稚園在園者数、小学校児童数、中学校生徒数、高等学校生徒数、短期大学学生数、大学学生数を対数処理しましょう。
 #まず対数処理をするまえにdescribeで概要をチェックする（正負など）
 #今回は負の値、もしくは０がないため、＋１の処理や符号処理やなどは行わない。

df["幼稚園在園者数"]=df["幼稚園在園者数"].map(lambda x:np.log(x))
df["小学校児童数"]=df["小学校児童数"].map(lambda x:np.log(x))
df["中学校生徒数"]=df["中学校生徒数"].map(lambda x:np.log(x))
df["高等学校生徒数"]=df["高等学校生徒数"].map(lambda x:np.log(x))
df["短期大学学生数"]=df["短期大学学生数"].map(lambda x:np.log(x))
df["大学学生数"]=df["大学学生数"].map(lambda x:np.log(x))

df.head(5)
```

```python
#問題１０
#ラベル名を編集しておく
df = df.rename(columns={"幼稚園在園者数": "幼稚園在園者数（対数）"})
df = df.rename(columns={"小学校児童数": "小学校児童数（対数）"})
df = df.rename(columns={"中学校生徒数": "中学校生徒数（対数）"})
df = df.rename(columns={"高等学校生徒数": "高等学校生徒数（対数）"})
df = df.rename(columns={"短期大学学生数": "短期大学学生数（対数）"})
df = df.rename(columns={"大学学生数": "大学学生数（対数）"})

df.head(5)
```

```python
#問題１１　データの分割　（各県の外国人比率の確認を目的としたデータ抽出を行う）
#各県の日本人の人口と外国人の人口を抽出したデータフレーム　df_2 を作りましょう
#⇒地域コード/地方カテゴリ/都道府県/日本人人口/外国人人口　がラベルになっているデータフレームを生成する。　


df_2=df[["地域コード","地方カテゴリ","都道府県","日本人人口","外国人人口"]].copy()

df_2.head(5)
```

```python
#問題１２
#df_2 に各県の外国人人口比率を算出した列を作りましょう
#外国人人口比率=外国人人口 ÷　（日本人人口＋外国人人口）×100

df_2['外国人人口比率 %']=(df_2['外国人人口']/(df_2['日本人人口']+df_2['外国人人口'])*100)
df_2

```

```python
#問題１３
#for文を使って、d2の地方カテゴリ別にデータを分割し、それぞれの地方カテゴリの名前のシートとして出力しましょう
#例　全国.csv 　/ 北海道.csv /東北.csv ・・・・etc

grouped=df_2.groupby("地方カテゴリ")
for name, data in grouped:
    data.to_csv(name + ".csv",index=False)
```
